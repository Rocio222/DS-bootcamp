```python
#Funcion para detectar strings en listas numericas:
def muestraLetras(df,columna):
    lista = set()
    for i in range(len(df)):
        try:
            df[columna].iat[i] = float(df[columna].iat[i]) 
        except:
            lista.add(df[columna].iat[i])
    return lista
```

```python
cols_int=["diag_1","diag_2","diag_3"]
df3[cols_int]=df3[cols_int]=df3[cols_int].astype('float64')
df3.shape
```

```python
#Divido el df en 2 (num y cat)
lista_numericas = df3._get_numeric_data().columns
df_num = df3[lista_numericas]
df_cat = df3.drop(lista_numericas, axis=1)
```

```python
#Imputacion de valores numericos
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
df_num_imp = imputer.fit_transform(df_num)
df_num = pd.DataFrame(df_num_imp, columns = df_num.columns)
df_num.isnull().values.any()
```

```python
#Imputacion del df categorico:
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df_cat_imp = imputer.fit_transform(df_cat)
df_cat = pd.DataFrame(df_cat_imp, columns = df_cat.columns)
df_cat.isnull().values.any()
```

```python
#Miro variabilidad en las cols:
for i in df_cat.columns:
    print(df_cat[i].value_counts(normalize=True))
    print('****************************')
```

```python
#Elimino cols sin variabilidad:
cols=["max_glu_serum","repaglinide","nateglinide","chlorpropamide","glimepiride","acetohexamide","glyburide",
     "tolbutamide","pioglitazone","rosiglitazone","acarbose","miglitol","troglitazone","tolazamide","examide",
     "citoglipton","glyburide-metformin","glipizide-metformin","glimepiride-pioglitazone","metformin-rosiglitazone",
     "metformin-pioglitazone"]
df_cat.drop(cols,axis="columns",inplace=True)
```

```python
#Hago subgrupos de categorias con porcentaje muy bajo:
df_cat['glipizide']=df_cat['glipizide'].replace(["Steady","Up","Down"],"Si")
df_cat['A1Cresult']=df_cat['A1Cresult'].replace([">8","Norm",">7"],"Si")
df_cat["age"]=df_cat["age"].replace(["[70-80)","[60-70)","[80-90)","[50-60)","[90-100)"],"Mayor50")
df_cat["age"]=df_cat["age"].replace(["[40-50)","[30-40)","[20-30)","[10-20)","[0-10)"],"Menor50")
df_cat["race"]=df_cat["race"].replace(["Hispanic","Asian"],"Other")
```

```python
#Creo variables Dummies:
dummy_insulin = pd.get_dummies(df_cat["insulin"])
dummy_insulin.rename(columns={'Down': 'Insuline_Down', 'Steady': 'Insuline_Steady','No': 'Insuline_No','Up': 'Insuline_Up' }, inplace=True)

dummy_age = pd.get_dummies(df_cat["age"])

dummy_gender= pd.get_dummies(df_cat["gender"])

dummy_metformin = pd.get_dummies(df_cat["metformin"])
dummy_metformin.rename(columns={'Down': 'metformin_Down', 'Steady': 'metformin_Steady','No': 'metformin_No','Up': 'metformin_Up' }, inplace=True)

dummy_glipizide = pd.get_dummies(df_cat["glipizide"])
dummy_glipizide.rename(columns={"Si":"glipizide_Si","No":"No_glip"},inplace=True)

dummy_change = pd.get_dummies(df_cat["change"])
dummy_change.rename(columns={"Ch":"change_Si","No":"No_change"},inplace=True)

dummy_A1Cresult = pd.get_dummies(df_cat["A1Cresult"])
dummy_A1Cresult.rename(columns={"Si":"Si_A1Cresult","No":"No_A1Cresult"},inplace=True)

dummy_race = pd.get_dummies(df_cat["race"])
```

```python
df_cat = pd.concat([df_cat,dummy_race,dummy_age,dummy_insulin,dummy_metformin,dummy_glipizide,dummy_change,dummy_A1Cresult], axis=1)
```

```python
#Normalizo los df_num
from sklearn.preprocessing import MinMaxScaler
df_num2 = MinMaxScaler().fit_transform(df_num)
df_num2 = pd.DataFrame(df_num2, columns = df_num.columns)
df_num2.head()
```

```python
#Transformo la col a predecir ('diabetesMed'):
from sklearn.preprocessing import LabelEncoder
class_labelEncoding = LabelEncoder()
clase = class_labelEncoding.fit_transform(df_cat['diabetesMed'].values)
clase
```

```python
#AÃ±ado diabetes_med numerica a df_cat--> clase
df_cat["clase"] = clase
```

```python
y = np.asarray(df_cat["clase"])
X = np.asarray(df_num2)
len(X)
```

```python
#Feature Selection
y = np.asarray(df_cat["clase"])
X = np.asarray(df_num2)

from sklearn.ensemble import ExtraTreesClassifier

# Build a forest and compute the feature importances
forest = ExtraTreesClassifier(n_estimators=250,random_state=0)
forest.fit(X, y)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), indices)
plt.xlim([-1, X.shape[1]])
plt.show()
```
