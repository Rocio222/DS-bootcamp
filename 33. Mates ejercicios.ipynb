### 5. Crea un programa que le pida a un usuario que ingrese una ecuaci√≥n de segundo grado. 

### Ingresa la informaci√≥n de la ecuaci√≥n como parametro de entrada a una funci√≥n. La funci√≥n tiene que devolver los valores de x1, x2 y la gr√°fica con la curva de la ecuaci√≥n, la recta y=0 y los dos puntos donde se interceptan.

```python
def ecuacion2(a,b,c):
    X = Symbol('X')
    ec=(a*X**2) + (b*X) +c
    ec_res=solve(ec)
    print(ec_res[0])
    ec1=int(ec_res[0])
    ec2=int(ec_res[1])
    x1=np.linspace(-20,20,50)
    y1=(a*x1**2)+(b*x1)+c
    plt.plot(x1, y1)
    plt.plot(x1,[0 for i in range(50)])
    plt.plot(ec1,0,marker="o", color="red")
    plt.plot(ec2,0,marker="o", color="red")
    plt.show()

    return(ec_res)

```

```python
int(ecuacion2(1,-5,-6)[0])
```

### 10. Crea una funci√≥n para encontrar la distancia entre dos puntos para  ùëõ  dimensiones.¬∂

```python
def distancia (x,y):
    if len(x)==len(y):
        distancia= np.linalg.norm(x-y)
        return distancia
    else:
        print("La longitudes no coinciden")
    
```

```python
x=np.array([-2,-3,-4])
y=np.array([3,4,2])
distancia(x,y)
```
