```python
# Ejecutar una query en varias lineas:
#Triple comillas!!!!
con= pymysql.connect(
host="localhost",
user="root",
password="696537626r",
db="employees",
cursorclass= pymysql.cursors.DictCursor)
cursor=con.cursor()
sql_query="""SELECT em.first_name,em.last_name,em.gender,CE.dept_no FROM employees.current_dept_emp CE
LEFT JOIN employees em
ON CE.emp_no=em.emp_no;"""
cursor.execute(sql_query)

data3=cursor.fetchall()
dt=pd.DataFrame(data3)
dt
```
