# Funci√≥n de Schwefel

```python
#Vamos a graficar la funci√≥n de schwefel
fig = plt.figure(figsize = (10,8))
ax = fig.gca(projection='3d')
# Valores del -100 al 100, con paso de 0.1
X = np.arange(-100, 100, 0.1)
Y = np.arange(-100, 100, 0.1)
X, Y = np.meshgrid(X, Y) #Creaci√≥n de la malla 

###########################  ECUACI√ìN ########################################
Z = 418.9829 * 2 - (X*np.sin(np.sqrt(abs(X)))+ Y*np.sin(np.sqrt(abs(Y)))) ##
##############################################################################

# Plot de la superficie
ax.plot_surface(X, Y, Z)
```

## 2. Funci√≥n McCormick 

```python
x_lista = (10,15,25,35,45)
y_lista = (15,20,30,40,50)

for x,y in zip(x_lista,y_lista):
    fx  = (math.sin(x+y)) + ((x-y)**2) - (1.5*x) + (2.5*y) +1
    print("%.2f"%fx)
```

## 3. Funci√≥n Bohachevsky 

```python
x_lista = (10,15,25,35,45)
y_lista = (15,20,30,40,50)
```

```python
for x,y in zip(x_lista,y_lista):
    fx=(x**2)+(2*y**2)-(0.3*math.cos(3*math.pi*x))-(0.4*math.cos(4*math.pi*y))+0.7
    print(fx)
```

## 4. Funci√≥n Drop-Wave

```python
for x,y in zip(x_lista,y_lista):
    a= 1+ (math.cos(12*math.sqrt(x**2+y**2)))
    b= (0.5*(x**2+y**2))+2
    fx= -(a/b)
    print(fx)
```

## 5. Funci√≥n Eggholder 

```python
for x,y in zip(x_lista,y_lista):
    a=(y+47)
    b=(math.sqrt(abs(y+(x/2)+47)))
    c=(math.sqrt(abs(x-(y+47))))
    fx=(-a*math.sin(b))-(x*math.sin(c))
    print(fx)
    
```

## 6. Funci√≥n Holder Table

```python
for x,y in zip(x_lista,y_lista):
    a=abs(1-((math.sqrt(x**2+y**2))/(math.pi)))
    b=math.sin(x)*math.cos(y)*math.exp(a)
    fx=-abs(b)
    print(fx)
    
```
