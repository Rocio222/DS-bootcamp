# Regresión Lógistica

La regresión lógistica es una técnica estadística y de machine learning que se utiliza para clasificar datos, mientras que las regresiones lineales, polinomial y no lineal buscan predecir una variable continua. **La regresión lógistica se enfoca en predicr la clase de una variable categorica.**

Por lo tanto forma parte de los modelos de clasificación. Es sencillo de implementar, tiene buen rendimiento cuando las clases son linealmente separables, trabaja mejor con clases binarias y es un modelo probabilístico. 

**La regresión lógistica mide la probabilidad de que un patrón pertenezca a una clase u otra.**

Para cada punto en el sistema se tiene que calcular el error y a partir de calcular todos los errores vamos a poder generar la función de coste

$$cost(\widehat{y},y) = \frac{1}{2}(\widehat{y}-y)^2$$

La función de coste es: 
    
$$J(W) = \frac{1}{n}\sum_{i=1}^{n}\frac{1}{2}(\widehat{y_i}-y_i)^2$$ 

- La ecuación de la recta azul la podemos representar como $-log(\widehat{y_i})$
- La ecuación de la recta verde la podemos representar como  $-log(1-\widehat{y_i})$

Por lo tanto para obtener el valor de $cost(\widehat{y},y)$ podemos utilizar la siguiente ecuación: 

$$cost(\widehat{y},y) = \left\{\begin{matrix}-log(\widehat{y_i}) \: \: \: \: si \: \: \: \:y= 1\\ -log(1-\widehat{y_i}) \: \: \: \: si \:\: \: \: y=0\end{matrix}\right.$$

- Podemos utilizar la ecuación anterior para enconrar el valor de $J(W)$:

$$J(W) = -\frac{1}{n}\sum_{i=1}^{n}y_i log(\widehat{y_i})+(1-y_i)log(1-\widehat{y_i})$$ 

#### Una vez encontrar la ecuación de coste ó *cost function*, el siguiente paso es minimizar su valor para todos los patrones, de esta forma estamos asegurando un mejor rendimiento de la clasificación

### Los pasos a seguir para la regresión lógistica

1. Inicializar los pesos $W$
2. Calcular $\widehat{y_i}$
3. Encontrar el error de todos los patrones de entrenamiento $J(W)$
4. Aplicar un método de optimización para minimizar $J(W)$ (gradiente descendente)
5. Actualizar los valores de los pesos $W$
6. Repetir el paso número 2.

### Regresión Lógistica con sklearn

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train, y_train)
yhat = clf.predict(X_test)
accuracy_score(y_test, yhat)
```
