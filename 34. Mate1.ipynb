# Ecuaci√≥n de una recta por dos puntos

La siguiente ecuaci√≥n representa la ecuaci√≥n de una recta: 
$$\frac{x-x_1}{x_2-x_1 } = \frac{y-y_1}{y_2-y_1 }$$

Si sustituimos los valores obtenemos: 
$$\frac{x-3}{(-4)-3 } = \frac{y-4}{(-3)-4 }$$

Desarrollamos la ecuaci√≥n
$$\frac{x-3}{-7} = \frac{y-4}{-7}$$ 
$$ x-3 = y-4 $$
$$ x-3+4 = y $$
$$ y = x+1 $$



```python
x = np.linspace(-5,4,20)
y = x+1
plt.plot(x, y)
plt.show()
```

# Ecuaci√≥n de dos rectas 

$$ x - y +1 = 0 $$
$$ 0.5x-y +1 = 0 $$

```python
x = np.linspace(-5,5,20)
y = x+1
y1 = 0.5*x+1
plt.plot(x, y, color= "green")
plt.plot(x,y1, color= "pink")
plt.show()
```

### Soluci√≥n del sistema de ecuaciones
- Despejar $x$ de la primera ecuaci√≥n
$$ x = y -1 $$
-Sustituir el valor de $x$ en la segunda ecuci√≥n
$$ 0.5 (y-1) -y + 1 = 0 $$
- Encontrar el valor de $y$
$$ 0.5y - 0.5 -y + 1 = 0 $$
$$ -0.5y +0.5 = 0 $$
$$y = \frac{-0.5}{0.5} = 1$$
- Encontrar el valor de $x$ a partir de $y$
$$x = 1-1 = 0 $$


```python
x = np.linspace(-5,5,20)
y = x+1
y1 = 0.5*x+1
plt.plot(x, y)
plt.plot(x,y1)
plt.plot(0,1,marker ='o', color="red")
plt.show()
```

# Distancia entre dos puntos 

```python
P1=(1,2)
P2=(4,5)
plt.plot(P1[0], P1[1], marker='o', markersize=6, color="red", label = "P1") #Punto  1
plt.plot(P2[0], P2[1], marker='o', markersize=6, color="blue", label = "P2") #Punto 2
plt.legend()
plt.show()
```

La distancia entre dos puntos esta dada por la siguiente ecuaci√≥n:

$$D(P1,P2) = \sqrt((x_2-x_1)^2 + (y_2-y_1)^2) $$

$$D(P1,P2) = \sqrt((4-1)^2+(5-2)^2)$$

$$D(P1,P2) = \sqrt(9+9)$$

$$ D(P1,P2) = 4.2426$$

La distancia entre los dos puntos esta dada por la siguiente ecuaci√≥n: 

$$D(P1,P2) = \sqrt((x_2-x_1)^2 + (y_2-y_1)^2 + (z_2-z_1)^2 $$

Si seguimos aumentando dimensiones, se agregan a la ecuaci√≥n para encontrar la distancia entre ambos puntos dependiendo de las dimensiones en las que se encuentren.

$$ P1 (x,y,z,k...n) $$
$$ P2 (x,y,z,k...n) $$

$$D(P1,P2) = \sqrt( x^2 + y^2 + z^2  + k^2 +...+ n^2 ) $$

A esta forma de encontrar la distancia entre dos puntos la podemos llamar: **distancia euclidiana**

La distancia de Chebysev es la mayor de las diferencias entre sus dimensiones.
