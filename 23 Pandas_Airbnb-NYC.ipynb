**6.	Crear una nueva columna con el rating de acuerdo a la siguiente formula, donde las variables son: puntaje del municipio, número de reviews y disponibilidad de la habitación al año.**

**Formula: Rating = (Puntaje de ciudad) + (número de review/ 629) + (disponibilidad / 365).**

El puntaje para cada ciudad es el siguiente {"Manhattan":10/10,"Brooklyn":8/10,"Queens":6/10,"Staten Island":4/10,"Bronx":2/10}.

```python
def rating(x,y,z):
    puntuaje={"Manhattan":10/10,"Brooklyn":8/10,"Queens":6/10,"Staten Island":4/10,"Bronx":2/10}
    a=puntuaje[x]
    return a + y/629 + z/365
    
```

```python
df["rating"]=df.apply(lambda x: rating(x["municipio"],x["number_of_reviews"],x["availability_365"]),axis=1)
```
