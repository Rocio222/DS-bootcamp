```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import RadiusNeighborsClassifier
from sklearn.neighbors import NearestCentroid
```

```python
df = pd.read_excel("chronickidney.xlsx")
```

```python
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
#Función dropna
df2 = df.copy()
#Si la columna tiene al menos 320 no-NaN se mantiene, todas las columnas con más de 80 NaNs se eliminan
df2.dropna(axis=1, thresh=320, inplace = True) 
df2.shape  
```

```python
lista_numericas = df._get_numeric_data().columns #utiliza dtype para identificar columnas númericas
```

```python
df_num = df[lista_numericas]
df_num.head(5)
```

```python
df_cat = df.drop(lista_numericas, axis=1)
df_cat.head(5)
```

```python
df_cat["al"] = df_num["al"]
del df_num["al"]
df_num.head(3)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
df_num_imp = imputer.fit_transform(df_num)
df_num = pd.DataFrame(df_num_imp, columns = df_num.columns)
df_num.head(10)
```

```python
df_cat.head(1)
```

```python
df_cat["pcv"]=df_cat["pcv"].replace("\t43",43)
df_cat["pcv"].unique()
df_cat["wc"]=df_cat["wc"].replace({"\t6200":6200,"\t8400":8400})
df_cat["wc"].unique()
df_cat["dm"]=df_cat["dm"].replace({"\tno":"no","\tyes":"yes"})
df_cat["dm"].unique()
df_cat["cad"]=df_cat["cad"].replace({"\tno":"no","\tyes":"yes"})
df_cat["cad"].unique()
```

```python
for i in df_cat:
    print(i)
    print(df_cat[i].unique())
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df_cat_imp = imputer.fit_transform(df_cat)
df_cat = pd.DataFrame(df_cat_imp, columns = df_cat.columns)
df_cat.head(10)
```

```python
for i in df_cat.columns.drop("class"):
    dum = pd.get_dummies(df[i], prefix=i)
    df_cat = pd.concat([df_cat,dum], sort=False, axis=1)
    del df_cat[i]
df_cat
```

```python
del df_cat["class"]
df_cat.head(1)
```

```python
df2 = pd.concat([df_cat,df_num], sort=False, axis=1)
df2.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
class_labelEncoding = LabelEncoder()
clase = class_labelEncoding.fit_transform(df["class"].values)
```

```python
X = np.array(df2)
y = clase
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
clasificador = NearestCentroid()
clasificador.fit(X_train,y_train)
y_pred = clasificador.predict(X_test)
clasificador.score(X_test,y_test)
```

```python
list_accuracy=[]
for i in range(3,50,2):
    clasificador = KNeighborsClassifier(i, algorithm="brute")
    clasificador.fit(X_train, y_train)
    yhat = clasificador.predict(X_test)
    resultado = accuracy_score(y_test, yhat)
    list_accuracy.append(resultado)
print(list_accuracy)
```

```python
np.argmax(list_accuracy)
```

```python
plt.plot(range(3,50,2), list_accuracy)
plt.ylabel('Accuracy ')
plt.xlabel('Number of Neighbors (K)')
plt.tight_layout() #Ajusta automáticamente los parámetros de la subtrama
plt.show()
```
