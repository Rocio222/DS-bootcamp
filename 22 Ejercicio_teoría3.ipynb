### 1. Eliminar las columnas: default, balance, housing, loan, duration, campaign, pdays, previous, poutcome

```python
df.drop(["default","balance", "housing", "loan", "duration", "campaign", "pdays", "previous", "poutcome"], axis = 'columns', inplace=True)
df.head()
```

### 2. Aplicar map sobre la columna contact y cambiar cada elemento a espaÃ±ol

```python
df["contact"].unique()
```

```python
df["contact"] = df["contact"].map({'cellular':"movil",'telephone':"fijo",'unknown':"desconocido"})
df.head()
```

### 3. Aplicar applymap sobre las columnas job y marital y pasa cada elemento a mayusculas

```python
df[["job","marital"]]=df[["job","marital"]].applymap(lambda x:x.upper())
df.head()
```

### 10. Agrupa los patrones por job y obten la edad mÃ¡xima y minima de cada agrupaciÃ³n

```python
df2=df.groupby("job",as_index=False)["age"].aggregate([max,min])
df2
```
