## 4. Obtener la siguiente gráfica

```python
from mpl_toolkits.mplot3d import Axes3D
import math
fig = plt.figure()
ax = fig.gca(projection='3d')
# Valores del -100 al 100, con paso de 0.1
X = np.arange(-100, 100, 0.1)
Y = np.arange(-100, 100, 0.1)
X, Y = np.meshgrid(X, Y) #Creación de la malla 

#ECUACIÓN
Z = X**2 + 2*Y**2 - 0.3*np.cos(3*np.pi*X) - 0.4*np.cos(4*np.pi*Y)+0.7

# Plot de la superficie
ax.plot_surface(X, Y, Z, color="magenta")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import math
from matplotlib import cm


fig = plt.figure()
ax = fig.gca(projection='3d')

x = np.arange(-100, 100, 0.1)
y = np.arange(-100, 100, 0.1)
x, y = np.meshgrid(x, y) 

z = x**2+2*(y**2)-0.3*np.cos(3*np.pi*x)-0.4*np.cos(4*np.pi*y)+ 0.7

ax.plot_surface(x, y, z, cmap="rainbow")

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

#función de schwefel
fig = plt.figure()
ax = fig.gca(projection='3d')
# Valores del -100 al 100, con paso de 0.1
X = np.arange(-100, 100, 0.1)
Y = np.arange(-100, 100, 0.1)

X, Y = np.meshgrid(X, Y) #Creación de la malla 

#ECUACIÓN
Z = X**2+2*Y**2-0.3*np.cos(3*np.pi*X)-0.4*np.cos(4*np.pi*Y) + 0.7

# Plot de la superficie
my_col = cm.jet(Z/np.amax(Z))
ax.plot_surface(X, Y, Z,facecolors =my_col, linewidth=1, antialiased=False)
plt.show()
```

## 15. Agrupa por rangos de edad y crea una grafica de sectores 


- Niño < 11 años
- Joven <18 años
- Adulto < 50 años
- Mayor >= 50 años


```python
rangoedad=[]
for i in df["Age"]:
    if i < 11:
        rangoedad.append("Niño")
    elif i<18:
        rangoedad.append("Joven")
    elif i<50:
        rangoedad.append("Adulto")
    else:
        rangoedad.append("Mayor")
df["Rango edad"] = rangoedad
df.head()

```

```python
df["Rango_edad2"] = df['Age'].apply(lambda x: "Niño" if x<11 else "Joven" if x<18 else "Adulto" if x<50 else "Mayor")
df
```

```python
def rango_edad(x):
    if x<11:
        k = "Nino"
    elif x<18:
        k = "Joven"
    elif x<50:
        k ="Adulto"     
    else:
        k ="Mayor" 
    return k 
```

```python
df["Rango de edad3"] = df.apply(lambda x:rango_edad(x["Age"]),axis=1)
df
```
