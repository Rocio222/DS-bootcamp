## Cursor

Los m√©todos que estamos empleando obtienen informaci√≥n por p√°ginas. Para obtener m√°s informaci√≥n tenemos que usar `Cursor`.

Sin `Cursor`, si queremos obtener, por ejemplo, todos los tweets de un usuario lo podemos hacer de la siguiente manera:

```python
page = 1
while True:
    tuits = api.user_timeline('sanchezcastejon', page=page)
    for tuit in tuits:
        print(tuit.text)
        print("************")
    page += 1
    if page==5:
        break
```

`Cursor` permite obtener todos directamente:

```python
tuits=tweepy.Cursor(api.user_timeline,'adreanceja').items()
```

```python
for tuit in tuits:
    print(tuit.text)
```

## Dataframe con datos de Twitter

**Vamos a crear un `DataFrame` de `pandas` con datos de los tweets de NEOLAND.**

```python
neoland = api.get_user('NeolandStudio')
display_user_info(neoland)
```

**Se crea una lista anidada con todos los tuits de NEOLAND:**

```python
tuits=tweepy.Cursor(api.user_timeline, neoland.id).items()
```

```python
tuitsneoland = [[tuit.text,tuit.created_at,tuit.id, tuit.retweet_count,tuit.favorite_count] for tuit in tuits]
```

`weekday` me da el d√≠a de la semana:

0 Lunes
1 Martes
etc.

```python
dfneoland.date[0].weekday()
```

Para traducir, usamos un diccionario:

```python
semana = {0:'Lunes',1:'Martes',2:'Mi√©rcoles',3:'Jueves',4:'Viernes',5:'S√°bado',6:'Domingo'}
```

```python
semana[dfneoland.date[0].weekday()]
```

Puedo crear columnas nuevas de mi dataframe con el a√±o, el mes, el d√≠a de la semana y la hora.

```python
dfneoland['year']=dfneoland.date.apply(lambda x: x.year)
dfneoland['month']=dfneoland.date.apply(lambda x: x.month)
dfneoland['weekday']=dfneoland.date.apply(lambda x: semana[x.weekday()])
dfneoland['hour']=dfneoland.date.apply(lambda x: x.hour)
```

```python
dfneoland.head()
```
