# Nube de palabras (Word clouds)


"Word clouds" funcionan de manera simple: cuanto más aparece una palabra específica en una fuente de datos textuales (como un discurso, publicación de blog o base de datos), más grande y audaz aparece en la palabra en la nube.


El paquete, llamado `word_cloud` fue desarrollado por ** Andreas Mueller **.(https://github.com/amueller/word_cloud/).


```python
# install wordcloud
#!conda install -c conda-forge wordcloud
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image # tratamiento de imagenes
```

Guardamos el archivo alicia, que es la novela de alicia en el país de las maravillas

```python
alice_novel = open('alicia.txt', 'r',encoding="UTF-8").read()
```

```python
len(alice_novel)
```

Mostramos las STOPWORDS

```python
stopwords=STOPWORDS
stopwords
```

Crea un objeto de nube de palabras y genera una nube de palabras.

```python
alice_wc = WordCloud(background_color='white', max_words=len(alice_novel), stopwords=stopwords)

# El objeto de word cloud
alice_wc.generate(alice_novel)
```

Ahora que se ha creado la nube word, la visualizamos

```python
fig, ax = plt.subplots(figsize=(12,12))
ax.imshow(alice_wc)
ax.axis('off')
plt.show()
```

```python
stopwords.add('said')
stopwords.add('one') 
stopwords.add('know') 
stopwords.add('will') 
stopwords.add('must') 
stopwords.add('went') 
stopwords.add('us') 
stopwords.add('come') 
stopwords.add('see') 
stopwords.add('well') 
stopwords.add('now') 
stopwords.add('may') 
stopwords.add('came') 
stopwords.add('seemed') 
stopwords.add('even') 
stopwords.add('project')
stopwords.add('Gutenberg')

# Creamos de nuevo el objeto
alice_wc = WordCloud(background_color='white', max_words=len(alice_novel), stopwords=stopwords)
alice_wc.generate(alice_novel)

fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(alice_wc)
ax.axis('off')
plt.show()
```

Otra cosa interesante que puede implementar con el paquete word_cloud es superponer las palabras en una máscara de cualquier forma. Usemos una máscara de Alice y su conejo.

```python
alice_mask = np.array(Image.open('alice_mask.png'))   
```

```python
alice_mask.shape
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.imshow(alice_mask)
plt.axis('off')
plt.show()
```

```python
# Creamos de nuevo el objeto agregando la mascara
alice_wc = WordCloud(background_color="white", max_words=2000, mask=alice_mask,
               stopwords=stopwords, contour_width=2, contour_color='steelblue')
alice_wc.generate(alice_novel)


fig, ax = plt.subplots(figsize=(20,20))
#ax.imshow(alice_wc)
ax.imshow(alice_wc, interpolation='bilinear')
ax.axis('off')
alice_wc.to_file("imagenalicia.png")
plt.show()

```
