```python
municipio=[]
comunidad_autonoma=[]
for j in range(2,len(tablas[2].find_all("tr"))):
    for x,i in enumerate(tablas[2].find_all("tr")[j].find_all("td",align ="left")): #municipios de la fila 1 
        if x%2==0:
            municipio.append(i.text)
        else:
            if i.text.find("\xa0")>-1:
                comunidad_autonoma.append(i.text[:i.text.find("\xa0")])
            else:
                comunidad_autonoma.append(i.text)
```

```python
comunidad_autonoma
```

```python
municipio=[]
comunidad_autonoma=[]
poblacion=[]
for j in range(2,len(tablas[2].find_all("tr"))):
    for x,i in enumerate(tablas[2].find_all("tr")[j].find_all("td",align ="left")): #municipios de la fila 1 
        if x%2==0:
            municipio.append(i.text)
        else:
            if i.text.find("\xa0")>-1:
                comunidad_autonoma.append(i.text[:i.text.find("\xa0")])
            else:
                comunidad_autonoma.append(i.text)
    for k in tablas[2].find_all("tr")[j].find_all("td",align ="right"):
        poblacion.append(k.text[9:9+k.text[10:].find("&")])
```

```python
print(len(municipio))
print(len(comunidad_autonoma))
print(len(poblacion))
```

```python
import pandas as pd
df = pd.DataFrame()
df["municipio"] = municipio
df["CA"]=comunidad_autonoma
df["Poblacion"]=poblacion
df.head()
```
