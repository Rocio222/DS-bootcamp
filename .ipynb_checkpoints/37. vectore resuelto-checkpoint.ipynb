### 2. Muestra 3 vectores ortogonales entre si mismos

```python
plt.figure(figsize=(5,5))

#Parametros: coordenadas base y largo en eje x y y 
plt.arrow(0, 0, -2, 2, head_width=1, head_length=1, color="purple")
plt.arrow(0, 0, 2.0, 2.0, head_width=1, head_length=1, color="green")
plt.arrow(0, 0, -2.0, -2.0, head_width=1, head_length=1, color="magenta")

plt.grid()
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.show()
```

```python
from mpl_toolkits import mplot3d

xr  = np.linspace(0,5,80)
yr  = np.linspace(-5,0,80)
zr  = np.linspace(0,5,80)
xx = [0]*80
yy = [0]*80
zz = [0]*80


fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(xr,yy,zz)
ax.scatter3D(xx,yr,zz)
ax.scatter3D(xx,yy,zr)

ax.set_xlim([-5,5])
ax.set_ylim([-5,5])
ax.set_zlim([-5,5])
plt.show()
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d.proj3d import proj_transform

class Arrow3D(FancyArrowPatch):
    def __init__(self, x, y, z, dx, dy, dz, *args, **kwargs):
        super().__init__((0,0), (0,0), *args, **kwargs)
        self._xyz = (x,y,z)
        self._dxdydz = (dx,dy,dz)

    def draw(self, renderer):
        x1,y1,z1 = self._xyz
        dx,dy,dz = self._dxdydz
        x2,y2,z2 = (x1+dx,y1+dy,z1+dz)

        xs, ys, zs = proj_transform((x1,x2),(y1,y2),(z1,z2), renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        super().draw(renderer)
def _arrow3D(ax, x, y, z, dx, dy, dz, *args, **kwargs):
    '''Add an 3d arrow to an `Axes3D` instance.'''

    arrow = Arrow3D(x, y, z, dx, dy, dz, *args, **kwargs)
    ax.add_artist(arrow)


setattr(Axes3D,'arrow3D',_arrow3D)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_zlim(-5,5)

ax.arrow3D(0,0,0,
           0,0,4,
           mutation_scale=20,
           #arrowstyle="-|>",
           linestyle='dashed',
          fc="blue")
ax.arrow3D(0,0,0,
           0,-4,0,
           mutation_scale=20,
           #arrowstyle="-|>",
           linestyle='dashed',
           fc="purple")
ax.arrow3D(0,0,0,
           4,0,0,
           mutation_scale=20,
           #arrowstyle="-|>",
           linestyle='dashed',
          fc = "pink")
fig.tight_layout()
```

### 3. Suma los siguientes vectores y obten el vector resultante

```python
v1 = np.array([2,0])
v2 = np.array([0,2])
v3 = np.array([2,2])
v4 = np.array([4,1])
v5 = np.array([2,3.5])
v6 = np.array([1.5,5])
vr = v1+v2+v3+v4+v5+v6
vr
```

```python
plt.figure(figsize=(7,7))

#Parametros: coordenadas base y largo en eje x y y 
plt.arrow(0, 0, 2.0, 0, head_width=0.3, head_length=0.3, color="black")
plt.arrow(2, 0, 0, 2.0, head_width=0.3, head_length=0.3, color="black")
plt.arrow(2, 2, 2, 2.0, head_width=0.3, head_length=0.3, color="black")
plt.arrow(4, 4, 4, 1.0, head_width=0.3, head_length=0.3, color="black")
plt.arrow(8, 5, 2, 4.0, head_width=0.3, head_length=0.3, color="black")
plt.arrow(10, 9, 1.5, 4.5, head_width=0.3, head_length=0.3, color="black")
plt.arrow(0, 0, vr[0], vr[1], head_width=0.3, head_length=0.3, color="magenta")

plt.xlim(-1,15)
plt.ylim(-1,15)
plt.show()
```

### 5. Haz el producto punto de los siguientes vectores y encuentra el angulo entre los vectores

 $\overrightarrow{AB} = 3\widehat{i} + 4\widehat{j} $
 
 
 $\overrightarrow{CD} = -2\widehat{i} + 7.5\widehat{j} $
 

```python
np.dot(np.array([3,4]),np.array([-2,7.5]))
```

```python
alpha = math.acos((np.dot(np.array([3,4]),np.array([-2,7.5])))/(math.sqrt(((0-3)**2)+(0-4)**2)*math.sqrt(((0+2)**2)+(0-7.5)**2) ))
alpha
```

```python
alpha = math.degrees(alpha)
alpha
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d.proj3d import proj_transform

class Arrow3D(FancyArrowPatch):
    def __init__(self, x, y, z, dx, dy, dz, *args, **kwargs):
        super().__init__((0,0), (0,0), *args, **kwargs)
        self._xyz = (x,y,z)
        self._dxdydz = (dx,dy,dz)

    def draw(self, renderer):
        x1,y1,z1 = self._xyz
        dx,dy,dz = self._dxdydz
        x2,y2,z2 = (x1+dx,y1+dy,z1+dz)

        xs, ys, zs = proj_transform((x1,x2),(y1,y2),(z1,z2), renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        super().draw(renderer)
def _arrow3D(ax, x, y, z, dx, dy, dz, *args, **kwargs):
    '''Add an 3d arrow to an `Axes3D` instance.'''

    arrow = Arrow3D(x, y, z, dx, dy, dz, *args, **kwargs)
    ax.add_artist(arrow)


setattr(Axes3D,'arrow3D',_arrow3D)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_zlim(-5,15)
ax.arrow3D(0,0,0,
           -1,4,0,
           mutation_scale=20,
           #arrowstyle="-|>",
           linestyle='dashed',
          fc="blue")
ax.arrow3D(0,0,0,
           -2,-3,0,
           mutation_scale=20,
           #arrowstyle="-|>",
           linestyle='dashed',
           fc="purple")
ax.arrow3D(0,0,0,
           0,0,-11,
           mutation_scale=20,
           #arrowstyle="-|>",
           linestyle='dashed')
fig.tight_layout()
```
