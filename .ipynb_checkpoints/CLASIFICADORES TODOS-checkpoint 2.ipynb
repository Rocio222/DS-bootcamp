# CLASIFICADORES TODOS

### Resumen de clasificadores:

- KNN (K nearest neigthbors)
- Clasificador euclideano.
- RadiusNN
- Naive Bayes
- Logistic Regression

### Validacion del modelo (Entrenamiento y testing)

- Hold out (La mas rapida, menos eficiente)
    + Dividir train y test
- Leave one out (La mas eficiente y mas lenta)
- Cross Validation o k-Fold (intermedia, la mas usada)

### Metricas de evaluacion del modelo

- Accuracy
- Jaccard index
- Precision
- Sensibilidad
- F1 Score

- Confussion Matrix: 
    + Evalua la precision de una clasificacion.
    + Muestra las predicciones correctas y equivocadas.
    + Muy util para ver verdaderos po y neg y falsos pos y neg
- Curva Roc
    + El area bajo la curva indica cuanto de correcta es la clasificacion.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import RadiusNeighborsClassifier
from sklearn.neighbors import NearestCentroid
from sklearn.model_selection import train_test_split
from sklearn.metrics import jaccard_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer
```

```python
df=pd.read_excel("/Users/rociocuestapedraz/Desktop/PYTHON bootcamp DS/datasets/chronickidney.xlsx")
```

#### Data wrangling

```python
#Reemplace("?" po Nan)
df.replace("?", np.nan, inplace = True)
```

```python
#Miro % de NANS
for col in df.columns:
    suma_nans=pd.isnull(df[col]).values.ravel().sum()
    print(col,"=",(suma_nans/df.shape[0])*100,"%")
    
```

```python
cols=["bp","pcc","ba","bu","sc","htn","dm","cad","appet","pe","ane","cad","htn","dm","sc","bu","ba","age","bgr"]
for i in cols:
    df=df.dropna(axis=0, subset=[i])
```

```python
#wc
df=df.replace('\t6200',6200)
df=df.replace('\t8400',8400)
#pcv
df=df.replace('\t43',43)
#dm
df=df.replace('\tyes',"yes")
#cad
df=df.replace('\tno',"no")
```

```python
df["wc"]=df["wc"].astype("float64")
df["pcv"]=df["pcv"].astype("float64")
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
class_labelEncoding = LabelEncoder()
y= class_labelEncoding.fit_transform(df["class"].values)
```

```python
df_num= df[['bp', 'sg', 'al', 'su','bgr', 'bu',
       'sc', 'sod', 'pot', 'hemo','rc','wc','pcv']]
df_cat= df[['rbc', 'pc', 'pcc', 'ba','htn', 'dm', 'cad',
       'appet', 'pe', 'ane']]
#y=df["class"]
df_cat=df_cat.reset_index(drop=True)
df_num=df_num.reset_index(drop=True)
```

```python
for col in df_cat.columns:
    suma_nans=pd.isnull(df_cat[col]).values.ravel().sum()
    print(col,"=",(suma_nans/df.shape[0])*100,"%")
```

### Variables categoricas:

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df_cat['rbc']= imputer.fit_transform(df_cat[['rbc']])
df_cat['pc']= imputer.fit_transform(df_cat[['pc']])
```

```python
#Transformar a dummys:
for category in df_cat:
        cat_dummies=pd.get_dummies(df_cat[category],prefix=category)
        df_cat=df_cat.join(cat_dummies)
        df_cat.drop([category], axis='columns', inplace=True)
```

```python
df_cat=df_cat[['rbc_normal','pc_normal','pcc_present',
 'ba_present','htn_yes','dm_yes','cad_yes','appet_good',
            'pe_yes','ane_yes']]
```

```python
df_cat
```

### Variables numericas

```python
for i in df_num:
    df_num[i]=df_num[i].replace(np.nan,df_num[i].mean())
```

```python
#Normalizar
from sklearn.preprocessing import MinMaxScaler
df_num2 = MinMaxScaler().fit_transform(df_num)
df_num2 = pd.DataFrame(df_num2, columns = df_num.columns)
df_num2.head()
```

```python
#Unir los dos dataframes
df2=df_cat.join(df_num2)
df2
```

```python
#Separar la clase del resto
y = np.asarray(y)
X = np.asarray(df2) 
```

```python
from sklearn.ensemble import ExtraTreesClassifier

# Build a forest and compute the feature importances
forest = ExtraTreesClassifier(n_estimators=250,random_state=0)
forest.fit(X, y)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), indices)
plt.xlim([-1, X.shape[1]])
plt.show()
```

## CLASIFICADOR KNN

## 1. Hold out

```python
#Divido en train y test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3)
```

```python
#Pruebo el mejor k de k-neigthboors
list_accuracy=[]
for i in range(3,50,2):
    clasificador = KNeighborsClassifier(i, algorithm="brute")
    clasificador.fit(X_train, y_train)
    yhat = clasificador.predict(X_test)
    resultado = accuracy_score(y_test, yhat)
    list_accuracy.append(resultado)
print(list_accuracy)
```

```python
#Ploteo Accuracy vs k
plt.plot(list_accuracy)
plt.ylabel('Accuracy ')
plt.xlabel('Number of Neighbors (K)')
plt.tight_layout() #Ajusta automáticamente los parámetros de la subtrama
plt.show()
```

```python
#Me quedo con el K que de mejor Accuracy
clasificador = KNeighborsClassifier(3, algorithm="brute")
clasificador.fit(X_train, y_train)
yhat = clasificador.predict(X_test)
resultado = accuracy_score(y_test, yhat)
resultado
```

```python
print("Jaccard index: " , jaccard_score(y_test, yhat, average="macro"))
print("Exactitud: " , accuracy_score(y_test, yhat))
print("Precisión: ", precision_score(y_test, yhat, average="macro"))
print("Sensibilidad: ", recall_score(y_test, yhat, average="macro"))
print("F1-score: ", f1_score(y_test,yhat, average="macro"))
```

### hold out confussion matrix

```python
#Utilizo y_test e y_hat
confusion_matrix(y_test, yhat, labels=[0,1])
```

```python
disp = plot_confusion_matrix(clasificador, X_test, y_test,
                             display_labels=[0,1,2],
                             cmap=plt.cm.Blues,
                             normalize="true")
plt.show()
```

```python
#Utilizo X_test e Y_test
probs = clasificador.predict_proba(X_test)
preds = probs[:,1]
fpr, tpr, threshold = roc_curve(y_test, preds)
roc_auc = auc(fpr, tpr)
print("Area bajo la curva: ",auc(fpr, tpr))
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.legend()
plt.ylabel('Sensibilidad')
plt.xlabel('1-Especificidad')
plt.show()
```

## 2. Leave one out

```python
#Para LOO no hay y_test, utilizamos la Y
from sklearn import neighbors
from sklearn.model_selection import LeaveOneOut
loo = LeaveOneOut()
yhat=[]
xx=0
for train_index, test_index in loo.split(X): #EL for va a realizar 150 iteraciones porq es un total de 150 patrones
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    #CLASIFICADOR 3NN
    clf = neighbors.KNeighborsClassifier(3)
    clf.fit(X_train, y_train)
    yhat1 = clf.predict(X_test)
    yhat.append(yhat1)
print("Accuracy: ", accuracy_score(y,yhat))
print("Jaccard index: " , jaccard_score(y, yhat, average="macro"))
print("Precisión: ", precision_score(y, yhat, average="macro"))
print("Sensibilidad: ", recall_score(y, yhat, average="macro"))
print("F1-score: ", f1_score(y,yhat, average="macro"))
```

```python
confusion_matrix(y, yhat, labels=[0,1])
```

```python
disp = plot_confusion_matrix(clf, X, y,
                             display_labels=[0,1],
                             cmap=plt.cm.Blues,
                             normalize="true")
plt.show()
```

```python
probs = clf.predict_proba(X)
preds = probs[:,1]
fpr, tpr, threshold = roc_curve(y, preds)
roc_auc = auc(fpr, tpr)
print("Area bajo la curva: ",auc(fpr, tpr))
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.legend()
plt.ylabel('Sensibilidad')
plt.xlabel('1-Especificidad')
plt.show()
```

## 3. Cross validation

### Busco el mejor k para Cross Validation:

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer

list_accuracy=[]
for i in range(3,50,2):
    clf = neighbors.KNeighborsClassifier(i)
    scores = cross_val_score(clf, X, y, cv=10, scoring = make_scorer(accuracy_score))
    list_accuracy.append(scores.mean())
print(list_accuracy)
```

```python
plt.plot(list_accuracy)
plt.ylabel('Accuracy ')
plt.xlabel('Number of Neighbors (K)')
plt.tight_layout() #Ajusta automáticamente los parámetros de la subtrama
plt.show()
```

```python
clf = neighbors.KNeighborsClassifier(3)
scores = cross_val_score(clf, X, y, cv=10, scoring = make_scorer(accuracy_score))
print(scores)
print("*****"*10)
print("Accuracy: ", scores.mean())
```

### Cross Validation Metrics

```python
clf = neighbors.KNeighborsClassifier(3)
scores_acc = cross_val_score(clf, X, y, cv=10, scoring = make_scorer(accuracy_score))
scores_Jac = cross_val_score(clf, X, y, cv=10, scoring = make_scorer(jaccard_score))
scores_prec = cross_val_score(clf, X, y, cv=10, scoring = make_scorer(precision_score))
scores_sens = cross_val_score(clf, X, y, cv=10, scoring = make_scorer(recall_score))
scores_f1=cross_val_score(clf, X, y, cv=10, scoring = make_scorer(f1_score))

#print(scores)
#print("*****"*10)
print("Accuracy_score: ", scores_acc.mean())
print("jaccard_score: ", scores_Jac.mean())
print("Precisión: ", scores_prec.mean())
print("Sensibilidad: ", scores_sens.mean())
print("F1-score: ", scores_f1.mean())
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10)
yhat=[]
ytest=[]
xtrain=[]
ytrain=[]
for train_index, test_index in kf.split(X):
    
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    clf = neighbors.KNeighborsClassifier(13)
    clf.fit(X_train, y_train)
    yhat1 = clf.predict(X_test)
    yhat.extend(yhat1)
    ytest.extend(y_test)
    xtrain.extend(X_train)
    ytrain.extend(y_train)
```

### Cross Validation Confussion matrix

```python
disp = plot_confusion_matrix(clf, xtrain, ytrain,
                             display_labels=[0,1],
                             cmap=plt.cm.Blues,
                             normalize="true")
plt.show()
```

### Cross Validation Curva Roc

```python
probs = clf.predict_proba(xtrain)
preds = probs[:,1]
fpr, tpr, threshold = roc_curve(ytrain, preds)
roc_auc = auc(fpr, tpr)
print("Area bajo la curva: ",auc(fpr, tpr))
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.legend()
plt.ylabel('Sensibilidad')
plt.xlabel('1-Especificidad')
plt.show()
```
