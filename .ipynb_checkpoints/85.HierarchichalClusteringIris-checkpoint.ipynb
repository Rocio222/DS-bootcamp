```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering  
```

```python
iris = datasets.load_iris()
X = iris.data[:,:2]
y = iris.target
```

```python
agglom = AgglomerativeClustering(n_clusters = 5, linkage = 'single')
agglom.fit(X[:,:2])
```

```python
agglom.n_clusters_ 
```

```python
agglom.labels_
```

```python
df = pd.DataFrame()
df["a1"]= X[:,0]
df["a2"] = X[:,1]
df["clase"]=agglom.labels_
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(df["a1"],df["a2"],hue = df["clase"], palette="Set2")
plt.show()
```

```python
agglom.n_leaves_
```

# DENDROGRAMA 

Es independiente al glommeringcluster 

```python
from scipy.spatial import distance_matrix 
dist_matrix = distance_matrix(X,X) 
print(dist_matrix)
```

```python
from scipy.cluster import hierarchy 
Z = hierarchy.linkage(dist_matrix, 'complete')
```

```python
import pylab
pylab.figure(figsize=(10,10))
dendro = hierarchy.dendrogram(Z)
```

# Ejemplo 2

```python
df = pd.read_csv("heart_failure_clinical_records_dataset.csv")
df.head()
```

```python
y = df["DEATH_EVENT"].values
```

```python
df = df.iloc[:,[7,4]]
df.head()
```

```python
X = df.values
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
X.shape
```

```python
agglom = AgglomerativeClustering(n_clusters = 2, linkage = 'complete')
agglom.fit(X[:,:2])
```

```python
agglom.labels_
```

```python
df = pd.DataFrame()
df["a1"]= X[:,0]
df["a2"] = X[:,1]
df["clase"]=agglom.labels_
```

```python
# en esta grafica se ven los puntos divididos de acuerdo a la columna de la clase real
plt.figure(figsize=(8,6))
sns.scatterplot(df["a1"],df["a2"],hue = y, palette="Set2")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(df["a1"],df["a2"],hue = df["clase"], palette="Set2")
plt.show()
```

```python
dist_matrix = distance_matrix(X,X) 
print(dist_matrix)
```

```python
Z = hierarchy.linkage(dist_matrix, 'complete')
dendro = hierarchy.dendrogram(Z)
```

```python
import pylab
pylab.figure(figsize=(18,50))
dendro = hierarchy.dendrogram(Z, leaf_rotation=0, leaf_font_size =12, orientation = 'right')
```
