```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/rociocuestapedraz/Desktop/PYTHON bootcamp DS/datasets/movie_review.csv')
df.head(5)
```

```python
#Para quitar simbolos raros
import re
def preprocessor(text):
    text = re.sub(r'<[^>]*>', '', text) #Para eliminar las etiquetas HTML
    text = re.sub(r'[\W]+', ' ', text.lower()) #Eliminamos todos los caracteres que no sean palabras
    return text
```

```python
df["text"]
```

```python
df['text'] = df['text'].apply(preprocessor)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

porter = PorterStemmer()

def tokenizer(text):
    return nltk.word_tokenize(text,"english")


def tokenizer_porter(text):
    return [porter.stem(word) for word in tokenizer(text)]
```

```python
df['text'] = df['text'].apply(tokenizer_porter)
```

```python
tokenizado=[]
for i in df['text']:
    texto=tokenizer_porter(i)
    tokenizado.append(texto)
```

```python
stop = nltk.corpus.stopwords.words("english")
token_limpio=[]
for token in tokenizado:
    lista_temp=[]
    token_limpio.append(lista_temp)
    for palabra in token:
        if palabra not in stop:
            lista_temp.append(palabra)
```

```python
lista=[]
for i in token_limpio:
    string=(' '.join(i))
    lista.append(string)
```

```python
from sklearn.preprocessing import LabelEncoder
class_labelEncoding = LabelEncoder()
y= class_labelEncoding.fit_transform(df["tag"].values)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

X = np.array(lista)
y = np.asarray(y)

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
print(X)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

tfidf = TfidfVectorizer()
pipe = Pipeline([('vect', tfidf), ('clf', LogisticRegression())])
pipe.fit(X_train,y_train)
pipe.score(X_test, y_test)
```
