# Vectores

Un vector es todo segmento de recta dirigido en el espacio.



Cada vector posee las siguientes características que son:

**Módulo**

Es la longitud o tamaño del vector. Para hallarla es preciso conocer el origen y el extremo del vector, para saber cuál es el módulo del vector, debemos medir desde su origen hasta su extremo.

El módulo de un vector $\overrightarrow{AB}$ es la longitud del segmento AB, se representa por $\left | \overrightarrow{AB} \right |$.



**Dirección**

Viene dada por la orientación en el espacio de la recta que lo contiene.

**Sentido**

Se indica mediante una punta de flecha situada en el extremo del vector, indicando hacia qué lado de la línea de acción se dirige el vector.

El sistema de referencia que se suele usar como norma general, es el Sistema de Coordenadas Cartesianas.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
plt.arrow(0, 0, 5, 5, head_width=1, head_length=1) # los primeros dos parametros son el origen, ek siguiente parametro es el largo del eje x 
# y el cuarto parametro el largo del eje y

plt.grid()
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

# Clases de vectores

### Vectores equipolentes
- Cuando tienen igual módulo, dirección y sentido.



```python
plt.figure(figsize=(5,5))

#Parametros: coordenadas base y largo en eje x y y 
plt.arrow(0, 0, 5.0, 5.0, head_width=1, head_length=1, color="b")
plt.arrow(0, 2, 5.0, 5.0, head_width=1, head_length=1, color="r")
plt.arrow(3, 4, 5.0, 5.0, head_width=1, head_length=1, color="purple")
plt.arrow(4, 0, 5.0, 5.0, head_width=1, head_length=1, color="green")

plt.grid()
plt.xlim(-2,10)
plt.ylim(-2,10)
plt.show()
```

### Vectores ligados
- Son vectores equipolentes que actúan en la misma recta ( tienen el mismo módulo, dirección, sentido y se encuentran en la misma recta)

```python
plt.figure(figsize=(5,5))

#Parametros: coordenadas base y largo en eje x y y 
plt.arrow(0, 0, 5.0, 5.0, head_width=1, head_length=1, color="b")
plt.arrow(6, 6, 5.0, 5.0, head_width=1, head_length=1, color="r")

plt.grid()
plt.xlim(-2,20)
plt.ylim(-2,20)
plt.show()
```

# Componentes de un vector

Cualquier vector puede ser considerado como resultado de la suma de vectores, cada uno de ellos en la dirección de uno de los ejes coordenados del sistema. 

$$\overrightarrow{AB} = x\overrightarrow{i}+y\overrightarrow{j}$$

Donde $x$ y $y$ son las magnitudes de los vectores auxiliares o componentes del vector $\overrightarrow{AB}$ y $\overrightarrow{i}$ y $ \overrightarrow{j}$ son los vectores unitarios de esas componentes

```python
plt.figure(figsize=(5,5))

#Parametros: coordenadas base y largo en eje x y y 
plt.arrow(0, 0, 2.0, 2.0, head_width=1, head_length=1, color="green")
plt.arrow(0, 0, 2.0, 0, head_width=1, head_length=1, color="black")
plt.arrow(0, 0, 0, 2.0, head_width=1, head_length=1, color="brown")

plt.grid()
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.show()
```

```python
plt.figure(figsize=(5,5))

#Parametros: coordenadas base y largo en eje x y y 
plt.arrow(0, 0, 2.0, 2.0, head_width=1, head_length=1, color="green")
plt.arrow(0, 0, 2.0, 0, head_width=1, head_length=1, color="black")
plt.arrow(2, 0, 0, 2.0, head_width=1, head_length=1, color="brown")

plt.grid()
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.show()
```

### Producto de un vector por un valor escalar

El resultado de multiplicar un valor escalar $k$ por un vector $\overrightarrow{AB}$, expresado analíticamente por $k *\overrightarrow{AB}$, es otro vector con las siguientes características :

1. Tiene la misma dirección que $\overrightarrow{AB}$.
2. Su sentido coincide con el de $\overrightarrow{AB}$, si $k$ es un número positivo, y es el opuesto, si $k$ es un número negativo.
3. El módulo es $k$ veces la longitud que representa el módulo de $\overrightarrow{AB}$ ( Si $k$ es 0 el resultado es el vector nulo).

$$ k * \overrightarrow{AB}$$

$$ k = 3 $$



```python
plt.figure(figsize=(5,5))
k=-3 #Como es negativo cambia el sentido del vector resultante

plt.arrow(0, 0, k*2, k*2, head_width=0.3, head_length=0.3, color="black")
plt.arrow(0, 0, 2, 2, head_width=0.3, head_length=0.3, color="red")

plt.grid()
plt.xlim(-7,7)
plt.ylim(-7,7)
plt.show()
```

### Producto escalar de dos vectores

Se obtiene de la suma de los productos formados por las componentes de uno y otro vector. Es decir, dados dos vectores $\overrightarrow{AB}$ y $\overrightarrow{CD}$, expresados en un mismo sistema de coordenadas:
$$\overrightarrow{AB} = x_{\overrightarrow{AB}}*\overrightarrow{i}+y_{\overrightarrow{AB}}*\overrightarrow{j}$$
$$\overrightarrow{CD} = x_{\overrightarrow{CD}}*\overrightarrow{i}+y_{\overrightarrow{CD}}*\overrightarrow{j}$$


$$\overrightarrow{AB}\cdot \overrightarrow{CD}= x_{\overrightarrow{AB}}* x_{\overrightarrow{CD}}+y_{\overrightarrow{AB}}*y_{\overrightarrow{CD}}$$

```python
plt.figure(figsize=(5,5))
plt.arrow(0, 0, 2, 4, head_width=0.3, head_length=0.3, color="green")
plt.arrow(0, 0, 2, 2, head_width=0.3, head_length=0.3, color="blue")

plt.grid()
plt.xlim(-1,4)
plt.ylim(-1,5)
plt.show()
```

```python
#Producto escalar 

AB = np.array([2,4])
CD = np.array([2,2])

np.dot(AB,CD)
```

Calcular el ángulo que hay entre dos vectores, es posible, ya que el producto escalar se puede hallar en función de sus módulos y del coseno del ángulo que forman mediante la siguiente fórmula :

$$ \overrightarrow{AB}\cdot \overrightarrow{CD} = \left | \overrightarrow{AB} \right | \cdot \left | \overrightarrow{CD} \right | \cdot cos (\alpha) $$

$$\alpha =  arcos (\frac{\overrightarrow{AB}\cdot \overrightarrow{CD}}{\left | \overrightarrow{AB} \right | \cdot \left | \overrightarrow{CD} \right |}) $$

```python
# producto de los modulos de AB y CD
np.linalg.norm(np.array([2,4])) * np.linalg.norm(np.array([2,2]))
```

### Producto vectorial entre dos vectores

El producto vectorial entre dos vectores, se define como un vector, donde su dirección es perpendicular al plano de los vectores, en el sentido del movimiento de un tornillo que gira hacia la derecha por el camino más corto del vector $\overrightarrow{AB}$ a $\overrightarrow{CD}$


$$ \overrightarrow{AB} \text{ x } \overrightarrow{CD} = \left | \overrightarrow{AB} \right | \cdot \left | \overrightarrow{CD} \right | \cdot sin (\alpha) $$




```python
np.linalg.norm(np.array([10,-3])) * np.linalg.norm(np.array([-2,4]))
```

```python
math.asin(34/46.69)
```

```python
math.degrees(0.8157026216464042)
```
