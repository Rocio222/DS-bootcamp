### Agregar una columna al dataframe inicial con el número total de migración de 1980 al 2013 por pais

```python
l4=[i for i in range(1980,2013)]+['Pais']
df6=df[l4].set_index('Pais').sum(axis=1).sort_values(ascending=False)
df['Total MIgracion']=df6.values
df.head()
```

### Encuentra el país de cada continente con más migración y crea un dataframe con estos paises

```python
continentes=df['Continente'].unique().astype('str')
paises=pd.DataFrame(columns=['Continente','Pais','Total Migracion'])
df_conta=df.groupby(['Continente','Pais'],as_index=False).sum()

for continente in continentes:
    df_cont=df_conta[df_conta['Continente']==continente]
    id_c=df_cont['Total MIgracion'].idxmax()
    x=df_cont.loc[id_c][['Continente','Pais','Total MIgracion']]

    paises=paises.append({'Continente':x['Continente'],'Pais':x['Pais'],'Total Migracion':x['Total MIgracion']},ignore_index=True)
paises
```

### Encontrar los 3 paises con más migración de cada continente y agregarlos en un dataframe

```python
continentes=df['Continente'].unique().astype('str')
paises=pd.DataFrame(columns=['Continente','Pais','Total Migracion'])
df_conta=df.groupby(['Continente','Pais'],as_index=False).sum()

for continente in continentes:
    df_cont=df_conta[df_conta['Continente']==continente]
    a=df_cont[['Continente','Pais','Total MIgracion']].sort_values(ascending=False,by='Total MIgracion').iloc[0:3]
    for i in range(min(3,len(a))):
        x=a.iloc[i]
        paises=paises.append({'Continente':(x['Continente']),'Pais':(x['Pais']),'Total Migracion':(x['Total MIgracion'])},ignore_index=True)
paises
```

### Cambiar los nombres de cada continente a español (utilizando un bucle for)

```python
lista=df.Continente.tolist()
lista_new=[]
for c in lista:
    if c=="Latin America and the Caribbean":
        lista_new.append("Sudamerica")
    elif c=="Northern America":
        lista_new.append("NorteAmerica")
    elif c=="Europe":
        lista_new.append("Europa")
    else:
        lista_new.append(c)

df["Continente"]=lista_new
df.head()
```

###  Obtener un dataframe que te parezca relevante a partir de la aplicación de filtros. ¡Se creativo!

```python
media_paises=(df.iloc[:,4:-1].mean(axis=1)).tolist()
std=(df.iloc[:,4:-1].std(axis=1)).tolist()
maximo=(df.iloc[:,4:-1].max(axis=1)).tolist()
minimo=(df.iloc[:,4:-1].min(axis=1)).tolist()

df3 =pd.DataFrame(columns = ["Continente","Pais","Total Migracion","Media","Maximo","Minimo","std"])
df3["Pais"]=df["Pais"]
df3["Continente"]=df["Continente"]
df3["Total Migracion"]=df["Total MIgracion"]
df3["Media"]=media_paises
df3["std"]=std
df3["Maximo"]=maximo
df3["Minimo"]=minimo
df3=df3.sort_values("Continente")
df3.reset_index(drop=True)

africa=df3[df3["Continente"]=="Africa"]
africa=africa.sort_values("Total Migracion",ascending=False)

sudamerica=df3[df3["Continente"]=="Sudamerica"]
sudamerica=sudamerica.sort_values("Total Migracion",ascending=False)

norteamerica=df3[df3["Continente"]=="Norteamerica"]
norteamerica=norteamerica.sort_values("Total Migracion",ascending=False)


europa=df3[df3["Continente"]=="Europa"]
europa=europa.sort_values("Total Migracion",ascending=False)

asia=df3[df3["Continente"]=="Asia"]
asia=asia.sort_values("Total Migracion",ascending=False)

oceania=df3[df3["Continente"]=="Oceania"]
oceania=oceania.sort_values("Total Migracion",ascending=False)

#pd.concat(africa,sudamerica,norteamerica,europa,asia,oceania)
asia
```
