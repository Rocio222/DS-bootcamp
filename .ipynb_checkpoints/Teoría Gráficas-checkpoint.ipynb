# Introducción a Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Para saber el tamaño de la figura por default 
plt.rcParams.get('figure.figsize')
```

```python
#Para cambiar el tamaño determinado
fig_size = plt.rcParams["figure.figsize"]  
fig_size[0] = 8 #Ancho de la gráfica 
fig_size[1] = 6 #Altura de la gráfica
plt.rcParams["figure.figsize"] = fig_size  
```

```python
x = np.linspace(-5, 5, 20)
y = x ** 2

plt.plot(x, y, color="green")  
plt.xlabel('Eje X')  
plt.ylabel('Eje Y')  
plt.title('Función cuadratica')  
plt.show()
```

```python
x = np.linspace(-5,5,20)
y = x**3
plt.plot(x,y, color = "green")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Función cubica")
plt.show()
```

```python
# subplot es una función que especifica la ubicación y el número de gráfica

x = np.linspace(-5, 5, 20)
y = x ** 2

#Como parametros de entrada para subplot vamos a tener 3: num filas, num de col, el num de la grafica
plt.subplot(2,2,1)  
plt.plot(x, y, color="green")  
plt.subplot(2,2,2)  
plt.plot(x, y, color="blue")  
plt.subplot(2,2,3)  
plt.plot(x, y, color="red")  
plt.subplot(2,2,4)  
plt.plot(x, y, color="magenta")  
plt.show()  
```

```python
# subplot es una función que especifica la ubicación y el número de gráfica
x = np.linspace(-5, 5, 20)
y = x

plt.subplot(2,3,1)  
plt.plot(x, y**1, color="green")  
plt.subplot(2,3,2)  
plt.plot(x, y**2, color="blue")  
plt.subplot(2,3,3)  
plt.plot(x, y**3, color="red")  
plt.subplot(2,3,4)  
plt.plot(x, y**4, color="magenta")  
plt.subplot(2,3,5)  
plt.plot(x, y**5, color="black")  
plt.subplot(2,3,6)  
plt.plot(x, y**6, color="brown")  
plt.show()  
```

```python
x = np.linspace(-5, 5, 20)
y = x

for k,i in enumerate(["green","blue","red","black","brown","magenta"]):
    plt.subplot(2,3,k+1)  
    plt.plot(x, y**(k+1), color=i)  
    plt.show()
```

```python
x = np.linspace(-5, 5, 20)
y = x

for k,i in enumerate(["green","blue","red","black","brown","magenta"]):
    plt.plot(x, y**(k+1), color=i)  
    plt.show()
```

```python
# De forma orientada a objetos

x = np.linspace(-10, 10, 20)
y = x ** 3


figure = plt.figure() #Creamos el objeto figura vacía 
axes = figure.add_axes([1,1, 1, 1])   #Asignamos el tamaño de los ejes
axes.plot(x, y, color='blue')  
axes.set_xlabel('Eje X')  
axes.set_ylabel('Eje Y')  
axes.set_title('Función cubica')  
plt.show()
```

```python
fig, ax = plt.subplots(3)
x = np.linspace(0, 10, 100)
ax[0].plot(x, np.sin(x))        
ax[1].plot(x, np.cos(x))
ax[2].plot(x, np.tan(x))
plt.show()
```

```python
x = np.linspace(-10, 10, 20)
y = x
figure, ax = plt.subplots(nrows=2, ncols=3)

k=1
for rows in ax:  
    for ax1 in rows:
        ax1.plot(x, y**k, 'b')
        k+=1
        
```

```python
# Para anañir leyendas

x = np.linspace(-2, 2, 20)
y = x ** 3
z = x ** 2
figure = plt.figure()
axes = figure.add_axes([1,1,1,1])
axes.plot(x, z, label="Función cuadrada", color="magenta")  
axes.plot(x, y, label="Función cubica")  
axes.legend(loc=2) 
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

#función de schwefel
fig = plt.figure()
ax = fig.gca(projection='3d')
# Valores del -100 al 100, con paso de 0.1
X = np.arange(-100, 100, 0.1)
Y = np.arange(-100, 100, 0.1)

X, Y = np.meshgrid(X, Y) #Creación de la malla 

#ECUACIÓN
Z = 418.9829 * 2 - (X*np.sin(np.sqrt(abs(X)))+ Y*np.sin(np.sqrt(abs(Y))))

# Plot de la superficie
ax.plot_surface(X, Y, Z, color="magenta")
plt.show()
```

```python
#Grafica de barras
etiquetas=["a","b","c","d"]
valores=[10,30,20,50]
plt.bar(etiquetas,valores, color=["blue","green","red","magenta"])
plt.show()
```

```python
etiquetas=["a","b","c","d"]
valores=[10,30,20,50]
plt.barh(etiquetas,valores, color="brown")
plt.show()
```

```python
fig, ax = plt.subplots()
etiquetas=["a","b","c","d"]
valores=[10,20,30,70]
valores1=[20,40,50,60]
x = np.arange(len(etiquetas))  # the label locations
ancho=0.4
plt.bar(x - ancho/2, valores, ancho, color="blue")
plt.bar(x + ancho/2, valores1, ancho ,color="red")
ax.set_xticks(x)
ax.set_xticklabels(etiquetas)
plt.show()
```

```python
fig, ax = plt.subplots()
etiquetas=["a","b","c","d"]
valores=[10,20,30,40]
valores1=[20,40,50,60]

x = np.arange(len(etiquetas))  # the label locations
ancho=0.5
plt.bar(etiquetas, valores, ancho, color="blue")
plt.bar(etiquetas, valores1, ancho ,color="red", bottom=valores)
ax.set_xticks(x)
ax.set_xticklabels(etiquetas)
plt.show()
```

```python
# Graficación en pila

años = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]

j1 = [8,10,17,15,23,18,24,29]  
j2 = [10,14,19,16,25,20,26,32]  
j3 = [12,17,21,19,26,22,28,10]

plt.plot([],[], color='magenta', label = 'jugador 1')  
plt.plot([],[], color='blue', label = 'jugador 2')  
plt.plot([],[], color='red', label = 'jugador 3')

plt.stackplot(años, j1, j2, j3, colors = ['magenta','blue','red'])  
plt.legend(loc=2)  
plt.title('Goles de los jugadores')  
plt.xlabel('año')  
plt.ylabel('Goles')  
plt.show()  
```

```python
# Grafica por sectores 
tipo_gol = 'Penalties', 'Gol de campo', 'tiro libre'
goles = [12,38,7]  
colores = ['magenta','green','red']
plt.pie(goles, labels = tipo_gol, colors=colores ,shadow = True, explode = (0.0, 0.1, 0.5), autopct = '%1.1f%%')  
plt.show()  
```
