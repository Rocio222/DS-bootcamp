# Graficación con seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df1 = pd.read_excel("Airbnb_NYC.xlsx",sheet_name = "NYC1")
df2 = pd.read_excel("Airbnb_NYC.xlsx",sheet_name = "NYC2")
df = pd.merge(df1,df2,right_on="id", left_on="id")
df.head(2)
```

```python
df.shape
```

```python
df = df.sort_values("price").iloc[1000:20000][["host_id","municipio","number_of_reviews","reviews_per_month","price","room_type"]]
df
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.pairplot(iris) #En la diagonal aparece el histograma y en los demás gráficas una gráfica de dispersión
plt.show()
```

```python
sns.pairplot(iris, hue="species") #Agregamos el parametro "hue" para diferenciar las clases 
plt.show()
```

```python
sns.pairplot(df, hue="room_type") #con la variable "hue" podemos ver cada categoría por color 
plt.show()
```

```python
sns.pairplot(df, vars=["price", "number_of_reviews"],height=5) #height nos va a servir para el tamaño de la gráfica
plt.show()
```

```python
sns.pairplot(df, vars=["price", "number_of_reviews"],height=3, diag_kind="kde") #kde nos muestra la densidad de una variable continua
plt.show()
```

```python
sns.pairplot(df, vars=["price", "number_of_reviews"],height=3, diag_kind="kde", markers="D") #Agregamos marcadores
plt.show()
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris.feature_names
```

```python
labels=np.array(iris.feature_names)
pattern=iris.data[140]
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
print(angles)
# Esto es para añadir el último lado del polígono:
pattern=np.concatenate((pattern,[pattern[0]]))
angles=np.concatenate((angles,[angles[0]]))


fig=plt.figure()
ax = fig.add_subplot(polar=True)
ax.plot(angles, pattern, 'o-', linewidth=4)
ax.fill(angles, pattern, alpha=1, color="yellow")
ax.set_thetagrids(angles * 180/np.pi, labels)
plt.show()
```
