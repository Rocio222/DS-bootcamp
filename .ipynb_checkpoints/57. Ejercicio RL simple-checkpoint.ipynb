```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from math import sqrt
```

```python
df=pd.read_csv("/Users/rociocuestapedraz/Desktop/PYTHON bootcamp DS/datasets/FuelConsumptionCo2.csv")
```

```python
df.head()
```

1. Prueba la regresi√≥n lineal con las siguientes variables: "CYLINDERS", "TRANSMISSION", "FUELCONSUMPTION_CITY", "FUELCONSUMPTION_HWY","FUELCONSUMPTION_COMB"

```python
df["TRANSMISSION"].unique()
```

```python
x=df["TRANSMISSION"].unique()
for i,j in zip(x,range(1,23)):
    df["TRANSMISSION"] = df["TRANSMISSION"].replace({i:j})
```

```python
df["TRANSMISSION"].unique()
```

```python
cols_X= ["CYLINDERS", "FUELCONSUMPTION_CITY", "FUELCONSUMPTION_HWY","FUELCONSUMPTION_COMB","TRANSMISSION"]

for i in cols_X:
    print("REGRESION PARA ",i,"vs CO2EMISSIONS")
    X= np.array(df[[i]])
    #X=X.reshape(-1, 1)
    y=np.array(df[['CO2EMISSIONS']])
    #y=y.reshape(-1, 1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
    regresion_lineal = linear_model.LinearRegression()
    regresion_lineal.fit(X_train, y_train)
    print ('a: ', regresion_lineal.coef_)
    print ('intercept: ',regresion_lineal.intercept_)
    
    sns.scatterplot(X_train.flatten(),y_train.flatten(), color="magenta") #entrenamiento
    sns.scatterplot(X_test.flatten(),y_test.flatten(), color="green") #prueba 
    plt.plot(X_train, regresion_lineal.coef_[0][0]*X_train + regresion_lineal.intercept_[0], '-b')
    plt.show()
    yhat = regresion_lineal.predict(X_test)
    from sklearn.metrics import mean_squared_error, r2_score
    mean_squared_error(yhat,y_test)
    score=r2_score(y_test,yhat)
    print("r-score = ",score)
    print("**********************************************")
    
```

2. Agrega la m√©trica de RMSE para evaluar las regresiones. (Programa la funci√≥n) 

```python
cols_X= ["CYLINDERS", "FUELCONSUMPTION_CITY", "FUELCONSUMPTION_HWY","FUELCONSUMPTION_COMB"]
for i in cols_X:    
    print("REGRESION PARA ",i,"vs CO2EMISSIONS")
    X= np.array(df[[i]])
    #X=X.reshape(-1, 1)
    y=np.array(df[['CO2EMISSIONS']])
    #y=y.reshape(-1, 1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
    regresion_lineal = linear_model.LinearRegression()
    regresion_lineal.fit(X_train, y_train)
    yhat = regresion_lineal.predict(X_test)
    from sklearn.metrics import mean_squared_error, r2_score
    RMSE=sqrt(mean_squared_error(yhat,y_test))
    print("RMSE: ", RMSE)
    print("*******************************************")
```
