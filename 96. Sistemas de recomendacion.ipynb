### K-Means

```python
X=np.array(df6)
inercias = [] 
  
for k in range(1,10): 
    kmeans = KMeans(k)
    kmeans.fit(X)     
    inercias.append(kmeans.inertia_)

plt.plot(range(1,10), inercias, 'bx-') 
plt.xlabel('Ks') 
plt.ylabel('Inercia') 
plt.show()
```

```python
#Numero optimo de clusters=4
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
df6["clase"]=kmeans.labels_
df6.head()
```
