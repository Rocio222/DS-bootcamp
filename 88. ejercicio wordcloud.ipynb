```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image # tratamiento de imagenes
```

```python
movidick= open('/Users/rociocuestapedraz/Desktop/movidick 2.txt','r', encoding='latin-1',errors='ignore').read()
```

```python
movidick
```

```python
len(movidick)
```

```python
stopwords=STOPWORDS
stopwords
```

```python
movidick2 = WordCloud(background_color='white', max_words=len(movidick), stopwords=stopwords)

# El objeto de word cloud
movidick2.generate(movidick)
```

```python
fig, ax = plt.subplots(figsize=(12,12))
ax.imshow(movidick2)
ax.axis('off')
plt.show()
```

```python
movi_mask=np.array(Image.open('/Users/rociocuestapedraz/Desktop/PYTHON bootcamp DS/scripts/alice_mask.png'))
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.imshow(movi_mask)
plt.axis('off')
plt.show()
```

```python
# Creamos de nuevo el objeto agregando la mascara
movidick2 = WordCloud(background_color="white", max_words=2000, mask=movi_mask,
               stopwords=stopwords, contour_width=2, contour_color='steelblue')
movidick2.generate(movidick)


fig, ax = plt.subplots(figsize=(20,20))
#ax.imshow(alice_wc)
ax.imshow(movidick2, interpolation='bilinear')
ax.axis('off')
movidick2.to_file("ballena.png")
plt.show()
```
