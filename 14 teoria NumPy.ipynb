## Números aleatorios con NumPy

```python
#Entre 0 y 1 y solo genera vectores
print(np.random.random())
print("*"*30)
print(np.random.random(15))
```

```python
#Entre 0 y 1
print(np.random.rand())
print("*"*50)
print(np.random.rand(3,3))
```

```python
#Distribución normal centrada en el cero con varianza igual a 1 
print(np.random.randn())
print("*"*30)
print(np.random.randn(2,2,2))
```

```python
#Enteros dado un rango
print(np.random.randint(0, 10))
print("*"*20)
print(np.random.randint(0, 10, size=[3,5]))
```

```python
#Similar a random.choices 
print(np.random.choice(['a', 'e', 'i', 'o', 'u'], size=10, p=[0.4, 0.1, 0.1, 0.2, 0.2])) 
```

```python
#Similar a random.seed()

for i in range(4):
    np.random.seed(4)
    print(np.random.randn())
```

```python
for i in range(4):
    rs = np.random.RandomState(15)
    print(rs.rand())
    print(rs.randint(0, 10))
    print(np.random.randn())
    print("*"*20)    
```
