**18. Agrega una nueva columna al daframe que se llame: "Rango edad" y va a catagorizar de la siguiente manera:**

- Mayor >= 50 aÃ±os

```python
lista=[]
for i in df.Age:
    if i<11:
        a="NiÃ±o"
        lista.append(a)
        
    elif 11<i<18:
        b="Joven"
        lista.append(b)
    elif 18<i<50:
        c="Adulto"
        lista.append(c)
    else:
        d="Mayor"
        lista.append(d)
df["Rango Edad"]=lista
df.head()
  
```
